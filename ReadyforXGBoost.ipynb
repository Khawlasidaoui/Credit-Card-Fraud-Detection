```python
import numpy as np
import pandas as pd
from pathlib import Path
from collections import Counter
```

```python
file_path = Path('card_transaction.v1.csv')
df = pd.read_csv(file_path)
df.head()
```

```python
df.columns = df.columns.str.replace(' ', '')
df
```

```python
df['Amount'] = pd.to_numeric(df['Amount'].str.replace('$', ''))
df
```

```python
cols_to_keep = ['Amount', 'Card', 'Year', 'Month', 'Day', 'MCC', 'IsFraud?']
```

```python
df['UseChip'].value_counts()
```

```python
usechip_dummies_df = pd.get_dummies(df['UseChip'])
```

```python
state_dummies_df = pd.get_dummies(df['MerchantState'])
state_dummies_df
```

```python
model_df = df[cols_to_keep]
model_df
```

```python
usechip_dummies_df.shape
```

```python
df.shape
```

```python
# the model that is ready for XGBoost and resampling
model_df = pd.concat([model_df, usechip_dummies_df, state_dummies_df], axis = 1)
model_df
```

```python
model_df['IsFraud?'].value_counts()
```

```python
#Looking for the amount of "yes" and 'no' fraudulent transactions for each year
for year in df['Year'].unique():
    print(year, df[df['Year']==year]['IsFraud?'].value_counts())
```

```python
# Change the stringd column
def change_string(Fraud):
    if Fraud == 'Yes':
        return 1
    else:
        return 0

model_df["IsFraud?"] = model_df["IsFraud?"].apply(change_string)
model_df.head()
```

```python
#looking for how many different card values there are
for card in model_df['Card'].unique():
    print(card, model_df[model_df['Card']==card]['Card'].value_counts())
```
